## Task #1: Vẽ đồ thị hồi quy tuyến tính

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Calculate linear regression (degree 1 polynomial)
slope, intercept = np.polyfit(x, y, 1)

# Create regression line
y_pred = slope * x + intercept

# Plot original data
plt.scatter(x, y)

# Plot regression line
plt.plot(x, y_pred)

# Labels
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression Example")

plt.show()
```

## Task #3: Gửi email bằng Python

```python
import smtplib
from email.message import EmailMessage

# Email details
sender_email = "nghipn.work@gmail.com"
receiver_email = "test@gmail.com"
password = "your_app_password"

# Create email
msg = EmailMessage()
msg["From"] = sender_email
msg["To"] = receiver_email
msg["Subject"] = "Email sent by ChatGPT"
msg.set_content("ChatGPT rocks!")

# Send email
with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
    server.login(sender_email, password)
    server.send_message(msg)

print("Email sent successfully!")
```

## Task #4: Scraping dữ liệu từ “Books to Scrape”

```python
import requests
from bs4 import BeautifulSoup

url = "https://books.toscrape.com/"

# Send request
response = requests.get(url)
response.raise_for_status()

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Find all books
books = soup.find_all("article", class_="product_pod")

for book in books:
    # Title
    title = book.h3.a["title"]
    
    # Price
    price = book.find("p", class_="price_color").text
    
    # Rating (stored in class name)
    rating = book.find("p", class_="star-rating")["class"][1]
    
    print(f"Title: {title}")
    print(f"Price: {price}")
    print(f"Rating: {rating}")
    print("-" * 40)
```

## Tạo Báo cáo Excel bằng Python

```python
from openpyxl import Workbook
from openpyxl.styles import Font
from openpyxl.utils import get_column_letter

# Create workbook and worksheet
wb = Workbook()
ws = wb.active
ws.title = "Sales Report"

# Header row
headers = ["Product", "Quantity", "Unit Price", "Total"]
ws.append(headers)

# Make headers bold
for cell in ws[1]:
    cell.font = Font(bold=True)

# Sample data
data = [
    ["Laptop", 10, 800],
    ["Mouse", 50, 20],
    ["Keyboard", 30, 45],
    ["Monitor", 15, 200],
]

# Add data rows with formula for Total column
for row_index, row in enumerate(data, start=2):
    ws.append(row)
    ws[f"D{row_index}"] = f"=B{row_index}*C{row_index}"

# Add grand total formula
last_row = ws.max_row
ws[f"C{last_row + 1}"] = "Grand Total:"
ws[f"C{last_row + 1}"].font = Font(bold=True)
ws[f"D{last_row + 1}"] = f"=SUM(D2:D{last_row})"
ws[f"D{last_row + 1}"].font = Font(bold=True)

# Auto-adjust column width
for col in ws.columns:
    max_length = 0
    col_letter = get_column_letter(col[0].column)
    for cell in col:
        if cell.value:
            max_length = max(max_length, len(str(cell.value)))
    ws.column_dimensions[col_letter].width = max_length + 2

# Save file
wb.save("Sales_Report.xlsx")

print("Excel report generated successfully!")
```
